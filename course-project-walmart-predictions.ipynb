### Training a decision tree model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
df_train
```

```python
preds = model.predict(X_test)
```

```python
dt_val_rmse = mean_squared_error(y_test, preds,squared=False)
```

```python
dt_train_rmse = mean_squared_error(model.predict(X_train), y_train,squared=False)
```

```python
print("Validation rmse {} and Train rmse {}".format(dt_val_rmse,dt_train_rmse))
```

```python
def test_params(**params):
    model = RandomForestRegressor(random_state=42, n_jobs=-1, **params).fit(X_train,y_train)
    rf1_train_preds = model.predict(X_train)
    rf1_val_preds = model.predict(X_test)
    train_rmse = mean_squared_error(rf1_train_preds,y_train,squared=False)
    val_rmse = mean_squared_error(rf1_val_preds,y_test,squared=False)
    return train_rmse, val_rmse
```

```python
def test_param_and_plot(param_name, param_values):
    train_errors, val_errors = [], [] 
    for value in param_values:
        params = {param_name: value}
        train_rmse, val_rmse = test_params(**params)
        train_errors.append(train_rmse)
        val_errors.append(val_rmse)
        #print(1, end = ' ')
    plt.figure(figsize=(10,6))
    plt.title('Overfitting curve: ' + param_name)
    plt.plot(param_values, train_errors, 'b-o')
    plt.plot(param_values, val_errors, 'r-o')
    plt.xlabel(param_name)
    plt.ylabel('RMSE')
    plt.legend(['Training', 'Validation'])
```

```python
test_param_and_plot('max_depth', [15,20,30,40])
```

The optimal depth of the trees will be 30.

```python
test_param_and_plot('max_features', ['auto','log2','sqrt'])
```

Clearly, 'auto' gives the least rmse. 

```python
test_param_and_plot('min_samples_split', [2,3,4,5])
```

The validation rmse is not changing significantly with min_sample_split value but training error is linearly increasing with min_sample_split value. So, it is better to stick to the default value i.e., 2.

```python
rf2 = RandomForestRegressor(random_state=42, n_jobs=-1, max_depth = 30, max_features='auto',min_samples_split=2)
```

```python
rf2.fit(X_train,y_train)
```

```python
rf2_train_preds = rf2.predict(X_train)
rf2_val_preds = rf2.predict(X_test)
train_rmse = mean_squared_error(rf2_train_preds,y_train,squared=False)
val_rmse = mean_squared_error(rf2_val_preds,y_test,squared=False)
```

```python
print('Validation RMSE: {}, Train RMSE: {}'.format(val_rmse, train_rmse))
```

The final random forest model is 1 unit better than the base random forest model.
